```python
#importing the libraries
import numpy as py
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv('insurance_pre.csv')
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)          #if male is '0' -then female is '1'
dataset                                                  #if smoker is '0'- then smoker is'1'  so use(drop_first=True)
```

```python
indep=dataset[['age', 'bmi', 'children','sex_male', 'smoker_yes']]
dep=dataset[['charges']]
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor

param_grid = {'criterion': ['mse', 'mae'],
              'n_estimators': [10, 100],
              'max_features': ['auto', 'log2']}

grid = GridSearchCV(RandomForestRegressor(), param_grid, refit=True, verbose=3, n_jobs=-1)
grid.fit(indep, dep)
```

```python
re=grid.cv_results_
```

```python


table=pd.DataFrame.from_dict(re)
```

```python
Future_prediction=grid.predict([[age_input,BMI_input,children_input,sex_male_input,Smoker_input]])
```
